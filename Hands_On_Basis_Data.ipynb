1. Install dan Jalankan MySQL

```python
!apt-get update
!apt-get install -y mysql-server
!service mysql start
```

2. Cek MySQL Sudah Aktif

```python
!mysql -u root -e "SELECT VERSION();"
```

3. Create Database

```python
!mysql -u root -e "CREATE DATABASE produk_db;"
```

4. Show Database

```python
!mysql -u root -e "SHOW DATABASES;"
```

5. Create Database

```python
!mysql -u root -D produk_db -e "CREATE TABLE produk (id_produk INT PRIMARY KEY, nama_produk VARCHAR(50), harga INT);"

#Menampilkan daftar tabel
!mysql -u root -D produk_db -e "SHOW TABLES;"

#Menampilkan struktur tabel produk
!mysql -u root -D produk_db -e "DESCRIBE produk;"
```

6. Insert Into

```python
!mysql -u root -D produk_db -e "INSERT INTO produk (id_produk, nama_produk, harga) VALUES (1, 'Laptop Asus', 8500000), (2, 'Mouse Logitech', 150000), (3, 'Keyboard Mechanical', 450000);"
```

7. Select

```python
!mysql -u root -D produk_db -e "SELECT * FROM produk;"
```

8. Update

```python
!mysql -u root -D produk_db -e "UPDATE produk SET harga = 450000 WHERE nama_produk = 'Laptop Asus';"

#Lihat hasilnya (SELECT)
!mysql -u root -D produk_db -e "SELECT * FROM produk;"

#Lihat data yg di update saja (SELECT)
!mysql -u root -D produk_db -e "SELECT * FROM produk WHERE nama_produk = 'Laptop Asus';"
```

9. Delete

```python
#Menghapus 1 data pada tabel awal misalnya id = 1
!mysql -u root -D produk_db -e "DELETE FROM produk WHERE nama_produk= ' Laptop Asus';"

#Memastikan hasilnya
!mysql -u root -D produk_db -e "SELECT * FROM produk;"
```
